```python
%matplotlib inline
def plot_alpha(Ts, betas_ref, betas_plot, name_ref, names_plot, filename=None):
    import matplotlib
    if filename is not None:
        matplotlib.use('Agg')
    import matplotlib.pyplot as plt

    fix, ax1 = plt.subplots()

    styles = ['k--', 'k-', 'b--', 'b-', 'g--', 'g-', 'r--', 'r-']
    i = 0

    print()
    for pressure, beta_ref in betas_ref.items(): 
        for bet_plot, name_plot in zip(betas_plot[pressure], names_plot[pressure]):
            alpha = bet_plot - beta_ref
            print("pressure", pressure, "name", name_plot, "style", styles[i]) 
            ax1.plot(Ts, alpha, styles[i], label=f"{name_plot} {pressure} GPa")
            i = i + 1

    ax1.set_ylabel(r"$\Delta^{}$Mg$_{}$ ".format('{26/24}', r'{X-\mathrm{'+name_ref+'}}') +
                   "(" + u'\u2030' + ")")
    ax1.set_xlabel("T (K)")
    ax1.axhline(0.0, color='k', linewidth=0.5)
    if filename is not None:
        plt.savefig(filename)
    else:
        plt.legend(ncols=2)
        plt.show()
```
